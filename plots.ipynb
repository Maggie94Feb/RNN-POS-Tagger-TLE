```python
import pandas as pd
df1 = pd.read_csv("bi-lstm_lr-0.5_training_stats.csv")
df2 = pd.read_csv("lstm_lr-0.5_training_stats.csv")

import matplotlib.pyplot as plt

plt.title('Loss Comparison (lr=0.5)')
plt.xlabel('epochs')
plt.ylabel('avg loss')
plt.grid()
plt.plot(df1["dev_loss"], "lime", label="bi-lstm_dev")
plt.plot(df1["train_loss"], "lightgreen", label="bi-lstm_train")
plt.plot(df2["dev_loss"], "red", label="lstm_dev")
plt.plot(df2["train_loss"], "salmon", label="lstm_train")
plt.legend()

plt.savefig("figures/lstm_lr-0.5_loss_comparison.png", dpi=200)
plt.show()

import matplotlib.pyplot as plt

plt.title('Accuracy Comparison (lr=0.5)')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.grid()
plt.plot(df1["dev_accuracy"], "lime", label="bi-lstm_dev")
plt.plot(df1["train_accuracy"], "lightgreen", label="bi-lstm_train")
plt.plot(df2["dev_accuracy"], "red", label="lstm_dev")
plt.plot(df2["train_accuracy"], "salmon", label="lstm_train")
plt.legend()

plt.savefig("figures/lstm_lr-0.5_accu_comparison.png", dpi=200)
plt.show()
```
